### Cleaning Data

```python
def clean_timestamp(df):
    # Find columns containing 'timestamp'
    timestamp_columns = [col for col in df.columns if 'timestamp' in col.lower()]

    if not timestamp_columns:
        raise ValueError("No timestamp column found in the DataFrame.")

    # Assume the first identified timestamp column is the correct one
    timestamp_column = timestamp_columns[0]

    # Convert the timestamp column to datetime type
    df[timestamp_column] = pd.to_datetime(df[timestamp_column])

    # Extract additional time-related features
    df['day_of_week'] = df[timestamp_column].dt.day_name()
    df['hour_of_day'] = df[timestamp_column].dt.hour

    return df

# Clean timestamp for df_day
try:
    df_day = clean_timestamp(df_day)
except ValueError as e:
    print(e)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
liburan_data = df_day[df_day["holiday"] == 1]

# Hitung jumlah total peminjaman sepeda saat liburan
total_peminjaman_liburan = liburan_data["cnt"].sum()

print("Jumlah total peminjaman sepeda saat liburan adalah:", total_peminjaman_liburan)
```

### Pertanyaan 2:

```python

# Hitung rata-rata peminjaman sepeda pada hari kerja dan akhir pekan
rata_rata_hari_kerja = df_day[df_day["workingday"] == 1]["cnt"].mean()
rata_rata_weekend = df_day[df_day["workingday"] == 0]["cnt"].mean()

print("Rata-rata peminjaman sepeda pada hari kerja:", rata_rata_hari_kerja)
print("Rata-rata peminjaman sepeda pada akhir pekan:", rata_rata_weekend)
```
